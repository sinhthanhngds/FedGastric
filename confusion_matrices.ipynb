```python
def plot_heatmap (cm, name, ax):
    sns.heatmap (cm, fmt = 'd', cmap = 'Greens', annot = True, ax = ax)
    ax.set_title (name)
    sns.set (font_scale = 1.9)
```

```python
def plot_plot (clins, title):
    fig, axs = plt.subplots (2, 2, figsize = (20, 15))
    plot_heatmap (clins['Clinic 0'], 'Clinic 0', axs [0, 0])
    plot_heatmap (clins['Clinic 1'], 'Clinic 1',axs [0, 1])
    plot_heatmap (clins['Clinic 2'], 'Clinic 2',axs [1, 0])
    plot_heatmap (clins['Clinic 3'], 'Clinic 3',axs [1, 1])
    fig.supxlabel ('Predicted Label')
    fig.supylabel ('True Label')
    fig.suptitle (title)
    plt.tight_layout()
```
